```python
# ============================================
#  Module 2 — Default KNN Baseline (No Scaling, No Tuning)
# ============================================

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(" Running Module 2: Default KNN Baseline (No Scaling, No Tuning)")

# ---------- 1. Load dataset ----------
df = pd.read_csv("/Users/ritawang/Desktop/learning/ESE 4170/Final/data/KNN/winequality-red_sep.csv")
df.columns = df.columns.str.strip()

X = df.drop(columns=["quality"])
y = df["quality"]

# ---------- 2. Train/Test split (Consistent with all modules) ----------
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("\n Train/Test shapes:", X_train.shape, X_test.shape)

# ---------- 3. Default KNN model ----------
knn_default = KNeighborsClassifier()   # default = n_neighbors=5, metric='minkowski', weights='uniform'

knn_default.fit(X_train, y_train)
y_pred = knn_default.predict(X_test)

# ---------- 4. Evaluation ----------
module2_acc = accuracy_score(y_test, y_pred)

print("\n===============================")
print(" Module 2 — Default KNN Baseline Results")
print("===============================")
print(f"Test Accuracy: {module2_acc:.4f}\n")

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\n Notes:")
print("- Default KNN performs poorly because features are not scaled.")
print("- Hyperparameters are not tuned.")
print("- Wine dataset has mixed units and strong class imbalance.")

# Save baseline results for Module 3
module2_best_params = {
    "n_neighbors": 5,
    "metric": "minkowski",
    "weights": "uniform",
    "p": 2
}

print("\n Saved for later modules:")
print("module2_acc =", module2_acc)
print("module2_best_params =", module2_best_params)

module2_acc
```

```python
# ============================================
#  Module 3 — Basic Hyperparameter Tuning
# ============================================

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

print("\n==============================")
print(" Module 3 — Basic Hyperparameter Tuning")
print("==============================\n")

# ---------------------------
# 1. Set the parameter adjustment space
# ---------------------------
param_grid = {
    "n_neighbors": [5, 11, 15, 19, 21, 25],
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

knn = KNeighborsClassifier()

grid = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

print(" Running GridSearchCV...")
grid.fit(X_train, y_train)

# ---------------------------
# 2.Output the best parameters and performance
# ---------------------------
module3_best_params = grid.best_params_
module3_cv_acc = grid.best_score_
module3_test_acc = grid.score(X_test, y_test)

print("\n Best Parameters Found:")
print(module3_best_params)

print(f"\nBest CV Accuracy: {module3_cv_acc:.4f}")
print(f" Test Accuracy AFTER tuning: {module3_test_acc:.4f}")

module3_acc = module3_test_acc

print("\nSaved module3_best_params and module3_acc for Module 4.")
```

```python
# ============================================
# Module 4 — Advanced Hyperparameter Tuning
# ============================================

from sklearn.model_selection import GridSearchCV

print("\n==============================")
print(" Module 4 — Advanced Hyperparameter Tuning")
print("==============================\n")



try:
    module3_best_params
except NameError:
    raise ValueError(" ERROR: module3_best_params NOT found. Run Module 3 first.")

print("Using best parameters from Module 3:")
print(module3_best_params, "\n")

base_k = module3_best_params["n_neighbors"]

# ---------------------------
# 1.Set the parameter adjustment space
# ---------------------------
param_grid_advanced = {
    "n_neighbors": [
        max(1, base_k - 6),
        base_k - 3,
        base_k,
        base_k + 3,
        base_k + 6
    ],
    "weights": ["uniform", "distance"],
    "metric": ["minkowski", "manhattan", "euclidean"],
    "p": [1, 1.5, 2, 3]  
}

knn_adv = KNeighborsClassifier()

grid_adv = GridSearchCV(
    estimator=knn_adv,
    param_grid=param_grid_advanced,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

print("Running Advanced GridSearchCV...\n")
grid_adv.fit(X_train, y_train)

# ---------------------------
# 2. Results
# ---------------------------
module4_best_params = grid_adv.best_params_
module4_cv_acc = grid_adv.best_score_
module4_test_acc = grid_adv.score(X_test, y_test)

print("\n Advanced Best Parameters:")
print(module4_best_params)

print(f"\n Module 4 CV Accuracy: {module4_cv_acc:.4f}")
print(f" Module 4 Test Accuracy: {module4_test_acc:.4f}")

module4_acc = module4_test_acc
print("\nSaved module4_acc for next Module.")
```

```python
# ================================
#  Module 5 – StandardScaler Optimization
# ================================

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier

print("\n===============================")
print(" Module 5 – StandardScaler Optimization")
print("===============================\n")

try:
    best_params_stage4 = module4_best_params
    print("Using module4_best_params from Module 4:")
except NameError:
    best_params_stage4 = module4_best_params
    print("Using advanced_best_params from Module 4 (no module4_best_params found):")

print(best_params_stage4, "\n")

module5_best_params = best_params_stage4.copy()


#  StandardScaler Pipeline
knn_scaled = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(
        n_neighbors=best_params_stage4["n_neighbors"],
        metric=best_params_stage4["metric"],
        weights=best_params_stage4["weights"],
        **({"p": best_params_stage4["p"]} if "p" in best_params_stage4 else {})
    ))
])

#  5-Fold Stratified CV
cv5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_m5 = cross_val_score(knn_scaled, X_train, y_train, cv=cv5)

module5_cv_acc = cv_scores_m5.mean()

print(" Module 5 CV Accuracy (with StandardScaler): {:.4f}".format(module5_cv_acc))
print("CV scores by fold:", cv_scores_m5, "\n")

# 4. TEST 
knn_scaled.fit(X_train, y_train)
y_test_pred_m5 = knn_scaled.predict(X_test)

module5_acc = accuracy_score(y_test, y_test_pred_m5)

print(" Module 5 Test Accuracy (with StandardScaler): {:.4f}".format(module5_acc))
print("\nClassification Report (Module 5):")
print(classification_report(y_test, y_test_pred_m5))
print("Confusion Matrix (Module 5):")
print(confusion_matrix(y_test, y_test_pred_m5))

# 5.Comparsion
try:
    print("\n=== Comparison with Module 4 ===")
    print("Module 4 Test Accuracy:", module4_acc)
    print("Module 5 Test Accuracy:", module5_acc)
    if module5_acc > module4_acc:
        print(" StandardScaler improved performance!")
    elif module5_acc < module4_acc:
        print(" Performance dropped slightly with StandardScaler.")
    else:
        print("ℹ No significant change.")
except NameError:
    print("\n( module4_acc not found, cannot compare automatically.)")


print("\nSaving key results for Module 6...")

saved_module5_acc = module5_acc
saved_module5_best_params = module5_best_params

print("✔ saved_module5_acc =", saved_module5_acc)
print("✔ saved_module5_best_params =", saved_module5_best_params)

```

```python
# ================================
#  Module 6 – Feature Selection (SelectKBest)
# ================================

from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier

print("\n===============================")
print(" Module 6 – Feature Selection Optimization")
print("===============================\n")

best_params_stage5 = module4_best_params  
print("Using previous best KNN parameters:")
print(best_params_stage5, "\n")

# k
k_values = [3, 4, 5, 6, 7, 8]

results = []

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for k in k_values:
    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("select", SelectKBest(f_classif, k=k)),
        ("knn", KNeighborsClassifier(
            n_neighbors=best_params_stage5["n_neighbors"],
            metric=best_params_stage5["metric"],
            weights=best_params_stage5["weights"],
            **({"p": best_params_stage5["p"]} if "p" in best_params_stage5 else {})
        ))
    ])

    cv_score = cross_val_score(pipe, X_train, y_train, cv=cv).mean()
    results.append((k, cv_score))

# cv-k
print("CV Accuracy for different K best feature numbers:")
for k, score in results:
    print(f"  k = {k}:  CV = {score:.4f}")

# best k
best_k, best_k_cv = max(results, key=lambda x: x[1])
print("\n Best k found:", best_k)
print(" Best CV accuracy with SelectKBest:", best_k_cv)

#retrain
final_pipe = Pipeline([
    ("scaler", StandardScaler()),
    ("select", SelectKBest(f_classif, k=best_k)),
    ("knn", KNeighborsClassifier(
        n_neighbors=best_params_stage5["n_neighbors"],
        metric=best_params_stage5["metric"],
        weights=best_params_stage5["weights"],
        **({"p": best_params_stage5["p"]} if "p" in best_params_stage5 else {})
    ))
])

final_pipe.fit(X_train, y_train)
y_pred_m6 = final_pipe.predict(X_test)

module6_acc = accuracy_score(y_test, y_pred_m6)

print("\n Module 6 Test Accuracy:", module6_acc)
print("\nClassification Report (Module 6):")
print(classification_report(y_test, y_pred_m6))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_m6))

# comparsion
print("\n===== Comparison with Module 5 =====")
print(f"Module 5 Test Accuracy: {module5_acc:.4f}")
print(f"Module 6 Test Accuracy: {module6_acc:.4f}")

if module6_acc > module5_acc:
    print("\n Feature Selection IMPROVED generalization. Keep as valid optimization.")
else:
    print("\n Feature Selection did NOT improve accuracy. Should be REJECTED in final report.")


module6_best_k = best_k
module6_acc = module6_acc
best_model_stage6 = final_pipe

print("\nSaved module6_acc =", module6_acc)
print("Selected features (k) =", best_k)



# 使用 Module 6 的最佳特征数量 k=6
selector = SelectKBest(score_func=f_classif, k=6)
selector.fit(X_train, y_train)

selected_mask = selector.get_support()
selected_features = X_train.columns[selected_mask]

print("👉 Selected features by SelectKBest (k=6):")
print(list(selected_features))
```

```python
# ============================================
#  Module 7 — PCA Feature Expansion Optimization
#      (based on Module 5 best model)
# ============================================

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd

print("\n============================================")
print("Module 7 — PCA Feature Expansion Optimization")
print("============================================\n")

print("Using best KNN parameters from Module 5:")
print(module5_best_params, "\n")
print(f"Module 5 Test Accuracy (baseline): {module5_acc:.4f}\n")

df = pd.read_csv("/Users/ritawang/Desktop/learning/ESE 4170/Final/data/KNN/winequality-red_sep.csv")
df.columns = df.columns.str.strip()

X = df.drop(columns=["quality"])
y = df["quality"]

# 3. Standardization (in the same way as Module 5
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 4.Generate two principal components of PCA as "new features"
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)

print("Explained variance ratio (PC1, PC2):", pca.explained_variance_ratio_, "\n")

# Piece the PCA features back to the original features (expand dimension, not reduce dimension)
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
X_expanded = pd.concat([X.reset_index(drop=True), df_pca], axis=1)

print("Original feature count:", X.shape[1])
print("New feature count after PCA expansion:", X_expanded.shape[1], "\n")

# 5. Build the KNN using the optimal parameters of Module 5
knn7 = KNeighborsClassifier(
    n_neighbors=module5_best_params["n_neighbors"],
    weights=module5_best_params["weights"],
    metric=module5_best_params["metric"],
    p=module5_best_params.get("p", 2)  
)

# 6. cv
cv_scores = cross_val_score(knn7, X_expanded, y, cv=5)
module7_cv = cv_scores.mean()

print(f"⭐ Module 7 CV Accuracy (with PCA expansion): {module7_cv:.4f}")
print("CV scores by fold:", cv_scores, "\n")

# 7.  Train/Test 
X_train, X_test, y_train, y_test = train_test_split(
    X_expanded, y, test_size=0.2, random_state=42, stratify=y
)

knn7.fit(X_train, y_train)
test_acc7 = knn7.score(X_test, y_test)

print(f" Module 7 Test Accuracy (with PCA expansion): {test_acc7:.4f}\n")

# 8. Comparsion
print(" Comparison with Module 5 baseline:")
print(f"  • Module 5 Test Accuracy: {module5_acc:.4f}")
print(f"  • Module 7 Test Accuracy: {test_acc7:.4f}\n")

if test_acc7 > module5_acc:
    print(" PCA dimensionality expansion enhances generalization ability, and this step can be marked as SUCCESSFUL optimization in the final report.")
else:
    print(" PCA dimensionality expansion did not bring about improvement. It can be attempted in the final report but is marked as REJECTED.")
    print(" The subsequent modules will continue to take Module 5 as the baseline.\n")

module7_acc = test_acc7          #  Test Accuracy
module7_best_params = module5_best_params  

print("Saved module7_acc =", module7_acc)
```

```python
# ================================
#  Module 8 – Mutual Information (MI) Feature Selection (非线性特征筛选)
# ================================

from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("\n===============================")
print(" Module 8 – Mutual Information (MI) Feature Selection")
print("===============================\n")

# ------------------------------------------------------
# 1.  Module 5 
# ------------------------------------------------------
try:
    base_params = best_params_stage5
    print("Using best KNN parameters from Module 5:")
except:
    base_params = best_params_stage4
    print("⚠️ Module 5 best params not found. Falling back to Module 4 parameters:")

print(base_params, "\n")

# ------------------------------------------------------
# 2.  top-k 
# ------------------------------------------------------
k_values = [3, 4, 5, 6, 7, 8, 9, 10]  
cv_results = {}
cv5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("CV Accuracy for different k (number of MI-selected features):\n")

for k in k_values:
    pipe = Pipeline([
        ("selector", SelectKBest(mutual_info_classif, k=k)),
        ("knn", KNeighborsClassifier(
            n_neighbors=base_params["n_neighbors"],
            metric=base_params["metric"],
            weights=base_params["weights"],
            **({"p": base_params["p"]} if "p" in base_params else {})
        ))
    ])

    scores = cross_val_score(pipe, X_train, y_train, cv=cv5)
    cv_results[k] = scores.mean()
    print(f"  k = {k}:  CV = {scores.mean():.4f}")

# ------------------------------------------------------
# 3.  k
# ------------------------------------------------------
best_k_mi = max(cv_results, key=cv_results.get)
best_cv_mi = cv_results[best_k_mi]

print("\n⭐ Best k selected by MI:", best_k_mi)
print("⭐ Best CV Accuracy:", best_cv_mi)

# ------------------------------------------------------
# 4.  test accuracy
# ------------------------------------------------------
final_model_mi = Pipeline([
    ("selector", SelectKBest(mutual_info_classif, k=best_k_mi)),
    ("knn", KNeighborsClassifier(
        n_neighbors=base_params["n_neighbors"],
        metric=base_params["metric"],
        weights=base_params["weights"],
        **({"p": base_params["p"]} if "p" in base_params else {})
    ))
])

final_model_mi.fit(X_train, y_train)
y_pred_m8 = final_model_mi.predict(X_test)

module8_acc = accuracy_score(y_test, y_pred_m8)

print("\n⭐ Module 8 Test Accuracy:", module8_acc)
print("\nClassification Report (Module 8):")
print(classification_report(y_test, y_pred_m8))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_m8))

# ------------------------------------------------------
# 5.  Module 5 baseline -compare
# ------------------------------------------------------
print("\n Comparison with Module 5 baseline:")
print("  • Module 5 Test Accuracy:", module5_acc)
print("  • Module 8 Test Accuracy:", module8_acc)

if module8_acc > module5_acc:
    print("\n MI-based Feature Selection IMPROVED accuracy → KEEP this module.")
else:
    print("\n MI Selection did NOT improve accuracy → REJECT this module in final report.")


module8_acc = module8_acc
best_model_stage8 = final_model_mi

print("\nSaved module8_acc =", module8_acc)
print("Selected MI top-k features =", best_k_mi)
```

```python
# ================================
#  Module 9 – Polynomial Feature Expansion (degree=2)
# ================================

from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("\n===============================")
print(" Module 9 – Polynomial Feature Expansion (degree=2)")
print("===============================\n")


try:
    best_params = module5_best_params 
    print("Using KNN parameters from Module 5:")
except NameError:
    best_params = best_params_stage5   
    print("Fallback: Using parameters from best_params_stage5:")
    
print(best_params, "\n")

baseline_acc = module5_acc
print(f"Module 5 baseline Test Accuracy: {baseline_acc:.4f}\n")

# ===========================================
# 2️Testing PolynomialFeatures + SelectKBest
# ===========================================

poly_knn_results = {}

print("🔍 Testing PolynomialFeatures + SelectKBest for different k values...\n")

for k in [10, 15, 20, 25, 30, 35, 40]:
    pipe = Pipeline([
        ("poly", PolynomialFeatures(degree=2, include_bias=False)),
        ("scaler", StandardScaler()),
        ("select", SelectKBest(score_func=f_classif, k=k)),
        ("knn", KNeighborsClassifier(
            n_neighbors=best_params["n_neighbors"],
            metric=best_params["metric"],
            weights=best_params["weights"],
            **({"p": best_params["p"]} if "p" in best_params else {})
        ))
    ])

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(pipe, X_train, y_train, cv=cv)

    poly_knn_results[k] = cv_scores.mean()
    print(f"k = {k}: CV Accuracy = {cv_scores.mean():.4f}")

# ===========================================
#  k
# ===========================================

best_k_poly = max(poly_knn_results, key=poly_knn_results.get)
best_cv_poly = poly_knn_results[best_k_poly]

print("\n Best k selected for PolynomialFeatures:", best_k_poly)
print(f" Best CV Accuracy (PolynomialFeatures): {best_cv_poly:.4f}\n")

# ===========================================
#  Test 
# ===========================================

final_poly_model = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("scaler", StandardScaler()),
    ("select", SelectKBest(score_func=f_classif, k=best_k_poly)),
    ("knn", KNeighborsClassifier(
        n_neighbors=best_params["n_neighbors"],
        metric=best_params["metric"],
        weights=best_params["weights"],
        **({"p": best_params["p"]} if "p" in best_params else {})
    ))
])

final_poly_model.fit(X_train, y_train)
y_test_pred_poly = final_poly_model.predict(X_test)

module9_acc = accuracy_score(y_test, y_test_pred_poly)

print(" Module 9 Test Accuracy:", module9_acc)
print("\nClassification Report (Module 9):")
print(classification_report(y_test, y_test_pred_poly))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_test_pred_poly))

# ===========================================
# compare Module 5 baseline 
# ===========================================

print("\n===============================")
print(" Comparison with Module 5 baseline")
print("===============================\n")
print(f"Module 5 Test Accuracy: {baseline_acc:.4f}")
print(f"Module 9 Test Accuracy: {module9_acc:.4f}")

if module9_acc > baseline_acc:
    print(" Polynomial Feature Expansion SUCCESSFULLY improved KNN performance!")
else:
    print(" Polynomial Feature Expansion did NOT improve accuracy → REJECT in final report.")



module9_best_params = best_params   
module9_best_k = best_k_poly  
module9_acc = module9_acc

print("\nSaved module9_best_params =", module9_best_params)
print("Saved module9_best_k =", module9_best_k)
print("Saved module9_acc =", module9_acc)
```

```python
# ================================
#  Module 10 – Feature Weighting Optimization
# ================================

from sklearn.feature_selection import f_classif
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np

print("\n===============================")
print("Module 10 – Feature Weighting Optimization")
print("===============================\n")

try:
    best_params_m9 = {
        "metric": best_params_stage5["metric"],
        "n_neighbors": module9_best_k,
        "weights": best_params_stage5["weights"],
        "p": best_params_stage5["p"]
    }
except:
    best_params_m9 = {"metric": "minkowski", "n_neighbors": 28, "weights": "distance", "p": 1}

print("Using KNN parameters from Module 9:")
print(best_params_m9, "\n")

# ---------------------------------------------------------
#  F-score
# ---------------------------------------------------------
F_values, p_values = f_classif(X_train, y_train)

weights = np.clip(F_values, a_min=0, a_max=None)
weights = weights / weights.max()

print("Raw F-score values:\n", F_values)
print("\nNormalized feature weights:\n", weights, "\n")

# ---------------------------------------------------------
# 3️Define the weighted distance function (based on Euclidean distance)
# ---------------------------------------------------------

def weighted_distance(x, y):
    return np.sqrt(np.sum(weights * (x - y)**2))

# sklearn 允许将 metric 传入自定义函数
def weighted_metric(a, b):
    return weighted_distance(a, b)

# ---------------------------------------------------------
# weighted distance - KNN
# ---------------------------------------------------------
knn_weighted = KNeighborsClassifier(
    n_neighbors=best_params_m9["n_neighbors"],
    metric=weighted_metric
)

# ---------------------------------------------------------
#  5-fold CV
# ---------------------------------------------------------
cv10 = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_m10 = cross_val_score(knn_weighted, X_train, y_train, cv=cv10)

module10_cv_acc = cv_scores_m10.mean()

print("⭐ Module 10 CV Accuracy (weighted KNN): {:.4f}".format(module10_cv_acc))
print("CV scores by fold:", cv_scores_m10, "\n")

# ---------------------------------------------------------
#  Fit + Test Accuracy
# ---------------------------------------------------------
knn_weighted.fit(X_train, y_train)
y_test_pred_m10 = knn_weighted.predict(X_test)

module10_acc = accuracy_score(y_test, y_test_pred_m10)

print("⭐ Module 10 Test Accuracy:", module10_acc)
print("\nClassification Report (Module 10):")
print(classification_report(y_test, y_test_pred_m10))
print("Confusion Matrix (Module 10):")
print(confusion_matrix(y_test, y_test_pred_m10))

# ---------------------------------------------------------
# Compare with Module 9 baseline 
# ---------------------------------------------------------
print("\n=== Comparison with Module 9 baseline ===")
print("Module 9 Test Accuracy:", module9_acc)
print("Module 10 Test Accuracy:", module10_acc)

if module10_acc > module9_acc:
    print(" Feature Weighting SUCCESSFULLY improved model performance!")
    result_m10 = "ACCEPTED"
elif module10_acc == module9_acc:
    print(" Feature Weighting produced SAME performance as Module 9.")
    result_m10 = "NEUTRAL"
else:
    print(" Feature Weighting REDUCED performance → REJECT this module.")
    result_m10 = "REJECTED"


module10_result = {
    "cv_acc": module10_cv_acc,
    "test_acc": module10_acc,
    "weights": weights,
    "status": result_m10
}

print("\nSaved module10_result =", module10_result)
```

```python
# ==============================================
#  Module 11 – Noise Cleaning (NCR) Optimization
# ==============================================
from imblearn.under_sampling import NeighbourhoodCleaningRule
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("\n=====================================")
print(" Module 11 – Noise Cleaning (NCR)")
print("=====================================\n")

try:
    print("Using best KNN parameters from Module 9:")
    print(module9_best_params, "\n")
    params_m11 = module9_best_params
except:
    print("Using manual fallback parameters (Module 9 not loaded).")
    params_m11 = {"n_neighbors": 35, "metric": "minkowski", "p": 1, "weights": "distance"}

baseline_acc = module9_acc  
print(f"Module 9 baseline Test Accuracy: {baseline_acc:.4f}\n")

print(" Applying Neighbourhood Cleaning Rule (NCR)...")

ncr = NeighbourhoodCleaningRule(n_neighbors=3)

X_train_cleaned, y_train_cleaned = ncr.fit_resample(X_train, y_train)

print(f"Original training size: {len(y_train)}")
print(f"Cleaned training size:  {len(y_train_cleaned)}")
print("\nClass distribution after NCR:")
print(y_train_cleaned.value_counts(), "\n")

knn_ncr = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(
        n_neighbors=params_m11["n_neighbors"],
        metric=params_m11["metric"],
        weights=params_m11["weights"],
        **({"p": params_m11["p"]} if "p" in params_m11 else {})
    ))
])

# ---------------------------
# 5-Fold CV on cleaned data
# ---------------------------
cv5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores_m11 = cross_val_score(knn_ncr, X_train_cleaned, y_train_cleaned, cv=cv5)

module11_cv_acc = cv_scores_m11.mean()
print(f"⭐ Module 11 CV Accuracy (after NCR): {module11_cv_acc:.4f}")
print("CV scores:", cv_scores_m11, "\n")

# ---------------------------
#  Test set evaluation
# ---------------------------
knn_ncr.fit(X_train_cleaned, y_train_cleaned)
y_test_pred_m11 = knn_ncr.predict(X_test)

module11_acc = accuracy_score(y_test, y_test_pred_m11)

print(f" Module 11 Test Accuracy: {module11_acc:.4f}\n")
print("Classification Report (Module 11):")
print(classification_report(y_test, y_test_pred_m11))
print("Confusion Matrix (Module 11):")
print(confusion_matrix(y_test, y_test_pred_m11), "\n")


print(" Comparison with Module 9 baseline:")
print(f"Module 9 Test Accuracy: {baseline_acc:.4f}")
print(f"Module 11 Test Accuracy: {module11_acc:.4f}")

if module11_acc > baseline_acc:
    print(" NCR SUCCESSFULLY improved generalization! Keep this module.")
    module11_status = "ACCEPTED"
else:
    print(" NCR did NOT improve accuracy → REJECT this module.")
    module11_status = "REJECTED"

module11_result = {
    "cv_acc": module11_cv_acc,
    "test_acc": module11_acc,
    "status": module11_status,
    "cleaned_size": len(y_train_cleaned)
}

print("\nSaved module11_result =", module11_result)
```

```python
# ================================
#  Module 12 – Multi-K Resampled Ensemble KNN
# ================================

from sklearn.ensemble import BaggingClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

print("\n===============================")
print(" Module 12 – Multi-K Resampled Ensemble KNN")
print("===============================\n")

base_params = module9_best_params
print("Using base parameters from Module 9:")
print(base_params, "\n")

p_val = base_params.get("p", 2)

K_list = [15, 25, 35, 45]
print("Multi-K ensemble will use K values:", K_list, "\n")

estimators = []

for k in K_list:
    knn = KNeighborsClassifier(
        n_neighbors=k,
        metric=base_params["metric"],
        p=p_val,
        weights=base_params["weights"]
    )

    bagged_knn = BaggingClassifier(
        estimator=knn,
        n_estimators=15,
        max_samples=0.8,
        bootstrap=True,
        random_state=42,
        n_jobs=-1
    )

    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("bagged_knn", bagged_knn)
    ])

    estimators.append((f"knn_k{k}", pipe))

# soft voting
ensemble_knn = VotingClassifier(
    estimators=estimators,
    voting="soft",
    n_jobs=-1
)

print("Training Multi-K Ensemble KNN...")
ensemble_knn.fit(X_train, y_train)

y_pred_m12 = ensemble_knn.predict(X_test)
module12_acc = accuracy_score(y_test, y_pred_m12)

print("\n Module 12 Test Accuracy:", module12_acc)
print("\nClassification Report (Module 12):")
print(classification_report(y_test, y_pred_m12))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_m12))
module12_result = {
    "acc": module12_acc,
    "K_list": K_list,
    "status": "SUCCESS"
}

print("\nSaved module12_result =", module12_result)
```
